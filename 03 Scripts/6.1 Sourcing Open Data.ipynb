## Data Wrangling

```python
# drop data we don't need
df = df.drop(['description',
              'facilities', 
              'picturecount',
              'houseNumber',
              'telekomTvOffer', 
              'telekomUploadSpeed', 
              'streetPlain',
              'scoutId',
              'firingTypes', 
              'pricetrend',
              'baseRentRange', 
              'noRoomsRange', 
              'thermalChar',
              'geo_plz',
              'yearConstructedRange',
              'livingSpaceRange',
             'regio2',
             'regio1'], axis=1)
```

```python
# Plot columns with missing values to evaluate how many are missing
fig, ax = plt.subplots()
(df.isnull().sum().sort_values(ascending=False)/df.shape[0]*100).sort_values(ascending=False).plot(kind='bar', figsize=(15, 10))
ax.set_xlabel("Columns")
ax.set_ylabel("Percentage of missing values")
ax.set_title("Columns with Percentage of Missing Values")
plt.show()
```

```python
# Columns with more than 50% missing values
df.columns[((df.isna().sum()/len(df)) > 0.50)]
```

```python
# delete columns with more than 50% misssing values
df = df.drop(columns=df.columns[((df.isna().sum()/len(df)) > 0.50)])
```

### Most important column for me is Total rent & livingspace, so let delete data with those values missing

```python
# drop rows with no total rent value 
df = df.drop(df[df['totalRent'] == 0.0].index)
```

```python
df.dropna(subset = ['totalRent'], inplace = True)
```

```python
# Drop rows with no living space
df = df.drop(df[df['livingSpace'] == 0.0].index)
```

```python
df.dropna(subset = ['livingSpace'], inplace = True)
```
